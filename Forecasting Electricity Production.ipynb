```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Simulate data for the past 10 years
years = np.arange(2012, 2022)
electricity_production = np.linspace(80_000, 106_114, len(years))  # GWh

df = pd.DataFrame({
    'Year': years,
    'Electricity_Production_GWh': electricity_production
})
df.set_index('Year', inplace=True)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Electricity_Production_GWh'], marker='o', linestyle='-')
plt.title('Simulated Electricity Production in the Philippines (2012-2021)')
plt.xlabel('Year')
plt.ylabel('Electricity Production (GWh)')
plt.grid(True)
plt.show()
```

```python
# Fit ARIMA model
model = ARIMA(df['Electricity_Production_GWh'], order=(1, 1, 1))
model_fit = model.fit()

# Forecast the next 5 years
forecast = model_fit.forecast(steps=5)
forecast_years = np.arange(2022, 2027)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Electricity_Production_GWh'], marker='o', label='Historical Data')
plt.plot(forecast_years, forecast, marker='x', linestyle='--', color='red', label='Forecast')
plt.title('Electricity Production Forecast for the Philippines')
plt.xlabel('Year')
plt.ylabel('Electricity Production (GWh)')
plt.legend()
plt.grid(True)
plt.show()
```
