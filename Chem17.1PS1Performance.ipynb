```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.style.use('ggplot')
pd.set_option('display.max_columns',200)
```

```python
df = pd.read_csv('17.1_PS1_Performance.csv')
#df = df.drop(columns=['Unnamed: 26','Unnamed: 33'], axis=1)
df.drop(columns=['Unnamed: 26','Unnamed: 33', 'Unnamed: 20'], axis=1, inplace=True)
df.rename(columns={'Unnamed: 0':'student'},inplace=True)
df.dropna(inplace=True)
df
```

```python
df.describe()
```

```python
# Strip spaces from column names
df.columns = df.columns.str.strip()

# Define max values safely
max_calo = df[['MC_Calo', 'TF_Calo']].max().sum()
max_uv_vis = df[['PS_UV-Vis', 'MC_UV-Vis', 'TF_UV-Vis']].max().sum()
max_equil = df[['MC_Equil', 'TF_Equil']].max().sum()
max_ph = df[['MC_pH', 'TF_pH']].max().sum()

# Handle missing columns gracefully
max_buffers = df.get(['MC_Buffers', 'TF_Buffers', 'PS_Buffer'], default=0).max().sum()
max_kinetics = df.get(['MC_Kinetics', 'TF_Kinetics', 'PS_Kinetics'], default=0).max().sum()

# Compute normalized values
df['calo'] = (df['MC_Calo'] + df['TF_Calo']) / max_calo * 100
df['UV-Vis'] = (df['MC_UV-Vis'] + df['TF_UV-Vis'] + df['PS_UV-Vis']) / max_uv_vis * 100
df['Equil'] = (df['MC_Equil'] + df['TF_Equil']) / max_equil * 100
df['pH'] = (df['MC_pH'] + df['TF_pH']) / max_ph * 100
df['Buffers'] = (df.get('MC_Buffers', 0) + df.get('TF_Buffers', 0) + df.get('PS_Buffer', 0)) / max_buffers * 100
df['Kinetics'] = (df.get('MC_Kinetics', 0) + df.get('TF_Kinetics', 0) + df.get('PS_Kinetics', 0)) / max_kinetics * 100
```

```python
df.head()
```

```python
df_summary = df[['calo','UV-Vis','Equil','pH','Buffers','Kinetics']].agg(['mean','std','min','max'])
print(df_summary)
```

```python
df_summary_mean = df_summary.corr()
sns.heatmap(df_summary_mean, annot=True, cmap='coolwarm')
plt.show()
```
