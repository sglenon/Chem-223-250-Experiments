# 8 | Liquid-Vapor Phase Diagram of Toluene-Benzene

```python
# I miswrote benzene as hexane oops haha but the code is too long na that i cant change, tnakakatmad

# constants
T = 304 #K
A_tol = 4.08245
B_tol = 1346.382
C_tol =-53.508
x_tol = np.linspace(0,1,11)
A_hex = 4.01814
B_hex = 1203.835
C_hex = -53.226
x_hex = 1 - x_tol

# Antoine Equation
ant_tol = A_tol - B_tol/(T+C_tol)
ant_hex = A_hex - B_hex/(T+C_hex)
P_star_tol = 10**ant_tol
P_star_hex = 10**ant_hex

# Vapor Pressure Calculations
P_tol = x_tol*P_star_tol
P_hex = x_hex*P_star_hex
P_total = P_tol + P_hex

# Mole fraction in vapor phase
y_tol = P_tol/P_total
y_hex = P_hex/P_total
```

```python
import pandas as pd

df = pd.DataFrame({
    "X(toluene)": x_tol,
    "X(hexane)": x_hex,
    "Y(toluene)": y_tol,
    "Y(hexane)": y_hex,
    "P(toluene)": P_tol,
    "P(hexane)": P_hex,
    "P(total)": P_total
})

df.head(11)
```

```python
# Plots
plt.plot(x_tol,P_total, label='Bubble Point Curve (liquid)', marker='o')
plt.plot(y_tol,P_total, label='Dew Point Curve (vapor)', marker='s')
plt.xlabel('Mole Fraction of Toluene')
plt.ylabel('Total Pressure (bar)')
plt.legend()
#plt.title('Binary Phase Diagram of a Toluene-Hexane System')
plt.show()
```

# IVB | Liquid-Vapor Phase Diagram of Hexane-Ethyl Acetate

```python
#constants
T = 298 #K

A_hex = 6.87776
B_hex = 1266.69
C_hex = 216.34
x_hex = np.linspace(0,1,11)

#Ethyl Acetate
A_ea = 7.1017
B_ea = 1277.03
C_ea = 237.23
x_ea = 1 - x_hex

# Antoine Equation
ant_ea = A_ea - B_ea/(T+C_ea)
ant_hex = A_hex - B_hex/(T+C_hex)
P_star_ea = 10**ant_ea
P_star_hex = 10**ant_hex

# Vapor Pressure Calculations
P_ea = x_ea*P_star_ea
P_hex = x_hex*P_star_hex
P_total = P_ea + P_hex

# Mole fraction in vapor phase
y_ea = P_ea/P_total
y_hex = P_hex/P_total
```

```python
import pandas as pd

df = pd.DataFrame({
    "X(ethyl acetate)": x_ea,
    "X(hexane)": x_hex,
    "Y(ethyl acetate)": y_ea,
    "Y(hexane)": y_hex,
    "P(ethyl acetate)": P_ea,
    "P(hexane)": P_hex,
    "P(total)": P_total
})

df.head(11)
```

```python
# Plots
plt.plot(x_hex,P_total, label='Bubble Point Curve (liquid)', marker='o',color='red')
plt.plot(y_hex,P_total, label='Dew Point Curve (vapor)',marker='s',color='orange')
plt.xlabel('Mole Fraction of Hexane')
plt.ylabel('Total Pressure (bar)')
plt.legend()
#plt.title('Binary Phase Diagram of a Toluene-Hexane System')
plt.show()
```

# IVA | Liquid-Solid Phase Diagram of an Fe-Al System

```python
# constants
dH_fe, Tf_fe = 13810, 1811
dH_al, Tf_al = 10710, 933.47
R = 8.314

#mole frax
x_fe = np.linspace(0.01,0.99,11)
x_al = 1 - x_fe

# den equations
den_fe = (1/Tf_fe) - (R * np.log(x_fe)) / dH_fe
den_al = (1/Tf_al) - (R * np.log(x_al)) / dH_al

# Temp
T_fe = 1 / den_fe
T_al = 1 / den_al
```

```python
df = pd.DataFrame({
    "X(Fe)": x_fe,
    "X(Al)": x_al,
    "T(Fe)": T_fe,
    "T(Al)": T_al})

df.head(11)
```

```python
# plot

plt.plot(x_fe, T_fe, label='iron', marker='o',color='red')
plt.plot(x_fe, T_al, label='aluminum', marker='s',color='orange')
plt.xlabel('Mole Fraction of Fe')
plt.ylabel('Temperature (K)')
plt.legend()


# Add vertical lines for pure substance melting points
plt.axvline(x=0, linestyle='--', color='gray', alpha=0.6)
plt.text(0.01, Tf_al + 20, "Pure Al\n933 K", color='orange')

plt.axvline(x=1, linestyle='--', color='gray', alpha=0.6)
plt.text(0.85, Tf_fe - 100, "Pure Fe\n1811 K", color='red')

# Estimate eutectic point from minimum of both curves
T_combined = np.minimum(T_fe, T_al)
index_eutectic = np.argmin(T_combined)
x_eut = x_fe[index_eutectic]
T_eut = T_combined[index_eutectic]

# Plot the eutectic point
plt.plot(x_eut, T_eut, 'ko')  # black dot
plt.text(x_eut + 0.02, T_eut + 20, f"Eutectic\n({x_eut:.2f}, {T_eut:.0f} K)", color='black')

# Annotate regions manually on the plot

# LIQUID region (above both liquidus lines)
plt.text(0.5, max(T_fe) + 50, "LIQUID", ha='center', fontsize=12, color='darkred')

# MIXTURE region (between the curves)
T_mid = 0.5 * (np.max(np.minimum(T_fe, T_al)) + np.min(np.maximum(T_fe, T_al)))
plt.text(0.5, T_mid, "SOLID + LIQUID", ha='center', fontsize=12, color='black')

# SOLID (Al-rich side)
plt.text(0.05, 700, "SOLID Al", ha='center', fontsize=11, color='blue')

# SOLID (Fe-rich side)
plt.text(0.95, 900, "SOLID Fe", ha='center', fontsize=11, color='red')


# You can include this if not yet done:
plt.xlabel("Mole Fraction of Iron (Fe)")
plt.ylabel("Temperature (K)")
#plt.title("SolidLiquid Phase Diagram (AlFe)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
T = [1500, 1600, 1700, 1800, 1900]
P = [19.72, 48.48, 107.2, 217.7, 408.2]

# Transform variables
y = np.log(P)
x = 1 / np.array(T)
x = x.reshape(-1, 1)  # Reshape to 2D array for sklearn

# Create and fit model
model = LinearRegression()
model.fit(x, y)

# Get slope, intercept, and R
slope = model.coef_[0]
m = slope
intercept = model.intercept_
r_squared = model.score(x, y)

# Output
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R: {r_squared}")

# dH
R = 8.314 #J/molK
dH = m*-1*R

print(dH)
```
