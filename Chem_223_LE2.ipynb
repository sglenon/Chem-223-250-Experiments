```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
E0_U = -0.577  # Standard potential for U4+/U3+
E0_Ce = 1.44   # Standard potential for Ce4+/Ce3+
n = 1          # electrons transferred per Ce4+/U3+ (1:1 reaction)

# Titration info
V_eq = 30.0  # mL at equivalence point
V_total = 40.0
V = np.linspace(0.01, V_total, 400)

# Total moles U and Ce
M_Ce = 0.015  # Molarity of Ce4+
mol_U = M_Ce * V_eq / 1000  # Since at equivalence, Ce = U

# Preallocate potential list
E = []

for v in V:
    mol_Ce_added = M_Ce * v / 1000

    if mol_Ce_added < mol_U:  # Before equivalence
        mol_U3 = mol_U - mol_Ce_added
        mol_U4 = mol_Ce_added
        ratio = mol_U4 / mol_U3
        E_val = E0_U + 0.05916 * np.log10(ratio)
    elif np.isclose(mol_Ce_added, mol_U, atol=1e-6):  # At equivalence
        E_val = (E0_U + E0_Ce) / 2
    else:  # After equivalence
        mol_Ce4 = mol_Ce_added - mol_U
        mol_Ce3 = mol_U
        ratio = mol_Ce4 / mol_Ce3
        E_val = E0_Ce + 0.05916 * np.log10(ratio)

    E.append(E_val)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(V, E, label="Titration Curve", color="darkblue")
plt.axvline(x=V_eq, color='red', linestyle='--', label="Equivalence Point")
plt.axhline(y=(E0_U + E0_Ce) / 2, color='green', linestyle=':', label="E at Equivalence")
plt.xlabel("Volume of Ce⁴⁺ added (mL)")
plt.ylabel("Cell Potential (V vs SHE)")
plt.title("Improved Redox Titration Curve: U³⁺ with Ce⁴⁺")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
