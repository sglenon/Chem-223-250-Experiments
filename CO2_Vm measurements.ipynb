# Cubic Equations of State and Pressure-Volume Isotherms

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
CO2_304 = pd.read_excel('CO2,304.13K.xlsx')
CO2_404 = pd.read_excel('CO2,404.13K.xlsx')
```

```python
#304.13K SRK

# important constants
R = 0.082057
T_c = 304.13
T = 304.13
P_c = 72.808
ω = 0.228

# SRK calculations for 304.13K
a = (0.42748*R**2*T_c**2)/(P_c)
m = (0.480)+(1.574*ω)-(0.1762*ω**2)
alpha = (1 + m*(1-(T/T_c)**(1/2)))**2
b = (0.08664*R*T_c)/(P_c)

# V_m calculations
V_m = CO2_304['Vm']
P = CO2_304['atm']
b_term = (V_m)**3 - (R*T/P)*(V_m)**2 - (b**2 + (b*R*T/P) - (a*alpha/P))*V_m - (a*alpha*b/P)
c_term = 3*(V_m)**2 - (R*T/P)*2*(V_m) - (b**2 + (b*R*T/P) - (a*alpha/P))
V_mx304 = V_m - (b_term/c_term)
```

```python
#404.13K SRK

# important constants
R = 0.082057
T_c = 304.13
T = 404.13
P_c = 72.808
ω = 0.228

# SRK calculations for 404.13K
a = (0.42748*R**2*T_c**2)/(P_c)
m = (0.480)+(1.574*ω)-(0.1762*ω**2)
alpha = (1 + m*(1-(T/T_c)**(1/2)))**2
b = (0.08664*R*T_c)/(P_c)

# V_m calculations
V_m = CO2_404['Vm']
P = CO2_404['atm']
b_term = (V_m)**3 - (R*T/P)*(V_m)**2 - (b**2 + (b*R*T/P) - (a*alpha/P))*V_m - (a*alpha*b/P)
c_term = 3*(V_m)**2 - (R*T/P)*2*(V_m) - (b**2 + (b*R*T/P) - (a*alpha/P))
V_mx404 = V_m - (b_term/c_term)
```

```python
#304.13K PR
# important constants
R = 0.082057
T_c = 304.13
T = 304.13
P_c = 72.808
ω = 0.228

# PR calculations for 404.13K
a = (0.45724*R**2*T_c**2)/(P_c)
k = (0.03764)+(1.5422*ω)-(0.26992*ω**2)
alpha = (1 + m*(1-(T/T_c)**(1/2)))**2
b = (0.07780*R*T_c)/(P_c)

#
V_m = CO2_304['Vm']
P = CO2_304['atm']
b_term = V_m**3 + (b-(R*T/P))*V_m**2 - (3*b**2 + (2*b*R*T/P)-(a*alpha/P))*V_m + (b**3 + (b**2*R*T/P)-(a*alpha*b/P))
c_term = 3*V_m**2 + 2*(b-(R*T/P))*V_m - (3*b**2 + (2*b*R*T/P)-(a*alpha/P))
V_my304 = V_m - (b_term/c_term)
```

```python
#404.13K PR
# important constants
R = 0.082057
T_c = 304.13
T = 404.13
P_c = 72.808
ω = 0.228

# PR calculations for 404.13K
a = (0.45724*R**2*T_c**2)/(P_c)
k = (0.03764)+(1.5422*ω)-(0.26992*ω**2)
alpha = (1 + m*(1-(T/T_c)**(1/2)))**2
b = (0.07780*R*T_c)/(P_c)

#
V_m = CO2_404['Vm']
P = CO2_404['atm']
b_term = V_m**3 + (b-(R*T/P))*V_m**2 - (3*b**2 + (2*b*R*T/P)-(a*alpha/P))*V_m + (b**3 + (b**2*R*T/P)-(a*alpha*b/P))
c_term = 3*V_m**2 + 2*(b-(R*T/P))*V_m - (3*b**2 + (2*b*R*T/P)-(a*alpha/P))
V_my404 = V_m - (b_term/c_term)
```

```python
#plot
plt.plot(V_mx304, P, label='304.13K SRK', color='red')
plt.plot(V_mx404, P, label='404.13K SRK', color='blue')
plt.plot(V_my304, P, label='304.13K PR', color='pink')
plt.plot(V_my404, P, label='304.13K PR', color='purple')
plt.plot(CO2_304['Vm'], CO2_304['atm'], label='304.13K exp', color='maroon')
plt.plot(CO2_404['Vm'], CO2_304['atm'], label='404.13K exp', color='black')
plt.xlabel('$V_m$')
plt.ylabel("P(atm)")
plt.title('Modelling of the Molar Volume-Pressure Relations of $CO_2$ using SRK and PR EOS')
plt.legend()
plt.show()
```
